## **Mission Control AI — Sistema Inteligente de Monitoramento Espacial**

> Sistema desenvolvido em Python utilizando Inteligência Artificial Generativa com Llama via Ollama para monitoramento de uma missão espacial. O sistema analisa temperatura, energia, comunicação e módulos, gerando alertas e recomendações automáticas.

### 1. Configuração do Ambiente e Modelo de Inteligência Artificial

> Nesta etapa é realizada a instalação do Ollama no Google Colab, inicialização do servidor e configuração do modelo Llama 3.2 1B utilizado para análise inteligente dos dados da missão.


```python
# 0. Instalar dependência necessária do Ollama no Colab

!apt-get update
!apt-get install -y zstd


# 1. Instalar o Ollama no Colab

!curl -fsSL https://ollama.com/install.sh | sh


# 2. Iniciar o servidor em background

import subprocess
import time

subprocess.Popen(["ollama", "serve"])

time.sleep(3)


# 3. Baixar o modelo Llama 3.1 1B

!ollama pull llama3.2:1b


# 4. Usar o modelo via biblioteca ollama

!pip install ollama -q

import ollama

resposta = ollama.chat(
    model="llama3.2:1b",
    messages=[
        {
            "role": "system",
            "content": "Você é um sistema de controle de missão espacial. Analise os dados e indique alertas."
        },
        {
            "role": "user",
            "content": "Temperatura: 95°C | Energia: 18% | Comunicação: instável. Avalie o status."
        }
    ]
)
print(resposta["message"]["content"])
```

### 2. Geração dos Dados Simulados da Missão

> O sistema gera automaticamente dados operacionais simulados de uma missão espacial, incluindo temperatura, nível de energia, comunicação e status dos módulos.

```python
import random

# Simulação dos sensores da missão

temperatura = round(random.uniform(0, 120), 1)
energia = random.randint(0, 100)
comunicacao = random.choice(
    [
        "Estável",
        "Instável",
        "Perdida"
    ]
)

modulo = random.choice(
    [
        "Operacional",
        "Manutenção",
        "Falha"
    ]
)

print("Dados coletados dos sensores:")
print(f"Temperatura: {temperatura}°C")
print(f"Energia: {energia}%")
print(f"Comunicação: {comunicacao}")
print(f"Status do módulo: {modulo}")
```

### 3. Sistema de Alertas e Tomada de Decisão

> Nesta etapa são aplicadas regras lógicas para identificar situações críticas e gerar alertas automáticos para auxiliar no controle da missão.



```python
alertas = []

# Análise da temperatura

if temperatura > 80:
    alertas.append(
        "Temperatura crítica detectada. Ativar resfriamento."
    )

# Análise de energia

if energia < 20:
    alertas.append(
        "Energia baixa. Ativar modo de economia."
    )

# Análise de comunicação

if comunicacao != "Estável":
    alertas.append(
        "Falha na comunicação detectada."
    )

# Análise do módulo

if modulo == "Falha":
    alertas.append(
        "Falha no módulo. Realizar manutenção."
    )

if len(alertas) == 0:
    alertas.append(
        "Todos os sistemas estão funcionando normalmente."
    )

print(" Alertas gerados:")

for alerta in alertas:
    print("-", alerta)
```

### 4. Análise Inteligente com IA Generativa
> Os dados coletados são enviados para o modelo de Inteligência Artificial Llama, que atua como um sistema de controle espacial responsável por interpretar os dados e recomendar ações.

```python
# Envio dos dados para a IA

dados_missao = f"""
Temperatura: {temperatura}°C
Energia: {energia}%
Comunicação: {comunicacao}
Status do módulo: {modulo}
Alertas do sistema:
{alertas}

Analise a situação da missão espacial e informe recomendações de ações para corrigir possíveis problemas.

"""

resposta = ollama.chat(
model="llama3.2:1b",
messages=[
{

"role": "system",
"content": "Você é uma inteligência artificial responsável pelo controle de uma missão espacial. Analise os dados, informe riscos e recomende ações corretivas."

},

{

"role":"user",
"content": dados_missao

}
]
)

# Salvando a análise da IA

analise_ia = resposta["message"]["content"]
```

### 5. Painel Mission Control — Visualização dos Dados

>Interface de visualização responsável por organizar as informações da missão, exibindo dados coletados, alertas do sistema e análise produzida pela IA.

```python
def exibir_status():
    print("=" * 50)
    print("MISSION CONTROL AI — STATUS DA MISSÃO")
    print("=" * 50)


    print(f" Temperatura: {temperatura}°C")
    print(f" Energia: {energia}%")
    print(f" Comunicação: {comunicacao}")
    print(f" Status do módulo: {modulo}")


    print("-" * 50)


    print(" ALERTAS DO SISTEMA:")


    for alerta in alertas:

        print("-", alerta)


    print("-" * 50)


    print(" ANÁLISE E RECOMENDAÇÕES DA IA:")

    print(analise_ia)


    print("=" * 50)



exibir_status()
```
